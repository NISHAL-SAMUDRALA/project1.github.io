```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset (final1.csv)
data = pd.read_csv('final1.csv')

# Assuming 'X' is the independent variable and 'Y' is the dependent variable
X = data['embed_0'].values.reshape(-1, 1)
y = data['embed_1'].values

# Split the data into training and testing sets (optional)
X_train = X[:-20]
X_test = X[-20:]
y_train = y[:-20]
y_test = y[-20:]

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error: %.2f" % mse)
# The coefficient of determination (R-squared): 1 is perfect prediction
r2 = r2_score(y_test, y_pred)
print("Coefficient of determination: %.2f" % r2)

# Plot outputs
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="blue", linewidth=3)

plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression')

plt.show()
```
