```python
#3
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

# Load your training dataset from a CSV file
# Replace 'your_dataset.csv' with the path to your training CSV file
data = pd.read_csv('final1.csv')

# Extract features (X_train) and labels (y_train) from the dataset
X_train = data.drop('embed_16', axis=1)  # Replace 'label_column_name' with the actual column name containing labels
y_train = data['Label']  # Replace 'label_column_name' with the actual column name containing labels

# Convert X_train and y_train to NumPy arrays
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()

# Create an SVM classifier
clf = svm.SVC()

# Fit the classifier to your training data
clf.fit(X_train, y_train)

# Load your test dataset from a CSV file
# Replace 'your_test_dataset.csv' with the path to your test CSV file
test_data = pd.read_csv('final1.csv')

# Extract features (X_test) and labels (y_test) from the test dataset
X_test = test_data.drop('embed_16', axis=1)  # Replace 'label_column_name' with the actual column name containing labels
y_test = test_data['Label']  # Replace 'label_column_name' with the actual column name containing labels

# Convert X_test and y_test to NumPy arrays
X_test = X_test.to_numpy()
y_test = y_test.to_numpy()

# Predict labels for the test data
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Generate a classification report
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```
